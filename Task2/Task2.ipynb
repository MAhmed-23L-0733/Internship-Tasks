```python
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Predicting Tesla stock prices by using Linear Regression model.


```python
ticker = "TSLA"
data = yf.download(ticker, start="2023-01-01", end="2024-12-31")
```

```python
data.head()
```

```python
X = data[["Open", "High", "Low", "Volume"]].values
```

```python
y = data["Close"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
lg = LinearRegression()
lg.fit(X_train, y_train)
```

Making predictions with the trained model.


```python
predictions = lg.predict(X_test)
```

Measuring the accuracy of the model.


```python
lg.score(X, y)
```

Plotting predicted vs actual measurements.


```python
plt.figure(figsize=(14, 7))
plt.plot(y_test, label="Actual Closing Price")
plt.plot(predictions, label="Predicted Closing Price")
plt.title(f"{ticker} Closing Price Prediction")
plt.xlabel("Time")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

Conclusion: The model is predicting the prices nearly similar to the actual closing prices with upto 99 percent accuracy.
