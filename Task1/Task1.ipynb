```python
x = np.array(iris["SepalLengthCm"])
y = np.array(iris["SepalWidthCm"])
```

```python
plt.scatter(x, y)
plt.xlabel("Sepal Length Cm")
plt.ylabel("Sepal Width Cm")
plt.show()
```

```python
x = np.array(iris["PetalLengthCm"])
y = np.array(iris["PetalWidthCm"])
```

```python
plt.scatter(x, y)
plt.xlabel("Petal Length Cm")
plt.ylabel("Petal Width Cm")
plt.show()
```

```python
x = np.array(iris["SepalLengthCm"])
plt.hist(x, edgecolor="black")
plt.xlabel("Sepal Length cm")
plt.ylabel("No")
plt.show()
```

```python
x = np.array(iris["SepalWidthCm"])
plt.hist(x, edgecolor="black")
plt.xlabel("Sepal Width cm")
plt.ylabel("No")
plt.show()
```

```python
x = np.array(iris["PetalLengthCm"])
plt.hist(x, edgecolor="black")
plt.xlabel("Petal Length cm")
plt.ylabel("No")
plt.show()
```

```python
x = np.array(iris["PetalWidthCm"])
plt.hist(x, edgecolor="black")
plt.xlabel("Petal Width cm")
plt.ylabel("No")
plt.show()
```

```python
x1 = np.array(iris["SepalLengthCm"])
x2 = np.array(iris["SepalWidthCm"])
x3 = np.array(iris["PetalLengthCm"])
x4 = np.array(iris["PetalWidthCm"])

data = [x1, x2, x3, x4]
plt.boxplot(
    data,
    vert=False,
    showmeans=True,
    labels=["Sepal Length Cm", "Sepal Width Cm", "Petal Length Cm", "Petal Width Cm"],
)
plt.show()
```
